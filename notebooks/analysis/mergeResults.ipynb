```python
import pandas as pd

# === Load Molecular Clock results for each cohort ===
INFORM = pd.read_csv("/Users/michail/HMS Dropbox/Michail Andreopoulos/HRDTimer/data/output/May27_TimingRun/INFORM_Breast_WGD_HRD_TimingResults_timing_nboot200_0001_only_prob_change_boot_pSub.csv")
PCAWG = pd.read_csv("/Users/michail/HMS Dropbox/Michail Andreopoulos/HRDTimer/data/output/May27_TimingRun/PCAWG_Breast_WGD_HRD_TimingResults_timing_nboot200_0001_only_prob_change_boot_pSub.csv")
SCANB = pd.read_csv("/Users/michail/HMS Dropbox/Michail Andreopoulos/HRDTimer/data/output/May27_TimingRun/SCANB_Breast_WGD_HRD_TimingResults_timing_nboot200_0001_only_prob_change_boot.csv")

# === Label cohorts ===
INFORM['cohort'] = 'INFORM'
PCAWG['cohort'] = 'PCAWG'
SCANB['cohort'] = 'SCANB'

# === Concatenate Molecular Clock results ===
combined_df = pd.concat([INFORM, PCAWG, SCANB], ignore_index=True)

# === Add sample.display from metadata ===
metadata = pd.read_csv("/Users/michail/HMS Dropbox/Michail Andreopoulos/HRDTimer/data/metadata/pan_metadata_v5.csv")
combined_df = combined_df.merge(
    metadata[['sample', 'sample.display']],
    left_on='ID', right_on='sample', how='left'
).drop(columns='sample')

# === Reorder columns ===
main_cols = ['ID', 'sample.display', 'cohort']
other_cols = [col for col in combined_df.columns if col not in main_cols]
combined_df = combined_df[main_cols + other_cols]

# === Load Age timing results ===
inform_age = pd.read_csv("/Users/michail/HMS Dropbox/Michail Andreopoulos/HRDTimer/data/output/Jun25_NewAgeModels/MC_to_Age_results_INFORM.csv")
pcawg_age = pd.read_csv("/Users/michail/HMS Dropbox/Michail Andreopoulos/HRDTimer/data/output/Jun25_NewAgeModels/MC_to_Age_results_PCAWG.csv")
scanb_age = pd.read_csv("/Users/michail/HMS Dropbox/Michail Andreopoulos/HRDTimer/data/output/Jun25_NewAgeModels/MC_to_Age_results_SCANB.csv")

# === Clean and prefix age columns ===
def prepare_age_df(df):
    df['ID'] = df['ID'].astype(str).str.strip()
    return df.rename(columns={col: f'Age_{col}' for col in df.columns if col not in ['ID', 'Age']})

inform_age = prepare_age_df(inform_age)
pcawg_age = prepare_age_df(pcawg_age)
scanb_age = prepare_age_df(scanb_age)

# === Merge only cohort-specific Age data ===
inform_merged = combined_df[combined_df['cohort'] == 'INFORM'].merge(inform_age, on='ID', how='left')
pcawg_merged  = combined_df[combined_df['cohort'] == 'PCAWG'].merge(pcawg_age, on='ID', how='left')
scanb_merged  = combined_df[combined_df['cohort'] == 'SCANB'].merge(scanb_age, on='ID', how='left')

# === Combine all back ===
final_combined_df = pd.concat([inform_merged, pcawg_merged, scanb_merged], ignore_index=True)

def merge_keep_unique_cols(main_df, right_df, key='ID'):
    # Find columns common to both dataframes, except the key column
    common_cols = set(main_df.columns).intersection(right_df.columns) - {key}
    # Drop those common columns from right_df before merge to avoid duplicates
    right_df_dropped = right_df.drop(columns=list(common_cols))
    # Merge
    merged_df = main_df.merge(right_df_dropped, on=key, how='left')
    return merged_df

# Load extra dfs
removal_df = pd.read_csv("/Users/michail/HMS Dropbox/Michail Andreopoulos/HRDTimer/data/output/sample_removal_reasons.csv").rename(columns={'sample_id': 'ID'})
removal_df['ID'] = removal_df['ID'].astype(str).str.strip()

pi1_df = pd.read_csv("/Users/michail/HMS Dropbox/Michail Andreopoulos/HRDTimer/data/output/All_Cohorts_pi1SBS3_vs_pi1SBS1_Jun11_noPi1_21.csv")
pi1_df['ID'] = pi1_df['ID'].astype(str).str.strip()

filters_df = pd.read_csv("/Users/michail/HMS Dropbox/Michail Andreopoulos/HRDTimer/data/output/HRDTimer_filters.csv").rename(columns={'Sample': 'ID'})
filters_df['ID'] = filters_df['ID'].astype(str).str.strip()

# Merge sequentially, keeping common columns only once
final_combined_df = merge_keep_unique_cols(final_combined_df, removal_df)
final_combined_df = merge_keep_unique_cols(final_combined_df, pi1_df)
final_combined_df = merge_keep_unique_cols(final_combined_df, filters_df)
final_combined_df

SBS1_SBS3_df = pd.read_csv("/Users/michail/HMS Dropbox/Michail Andreopoulos/HRDTimer/data/output/All_Cohorts_SBS1_SBS3_Late_Early_Jun11_noPi1_21.csv")
final_combined_df = final_combined_df.merge(SBS1_SBS3_df, on="ID", how="left")
final_combined_df.to_csv("All_data_Jun30.csv", index=False)

final_combined_df.columns
```
